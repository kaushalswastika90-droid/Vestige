```python
list1 = [11,22,44,22,11,33,66,77,88,11,44,55,33,11]
list2 = []

max_count = 0
max_element = None

for i in list1:
    if i not in list2:
        list2.append(i)
        count = list1.count(i)

        if count > max_count:
            max_count = count
            max_element = i

print("Element with maximum frequency:", max_element)
print("Count:", max_count)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = []

for i in range(len(list1)):
    product = 1
    for j in range(len(list1)):
        if i != j:
            product *= list1[j]
    list2.append(product)

print(list2)
```

```python
data = "cranes varsity bangalore"

unique = set(data)

repeated = set()

for ch in data:
    if data.count(ch) > 1:
        repeated.add(ch)

print("unique characters number:", len(unique))
print("unique characters list:", unique)
print("repeated characters:", repeated)
```

```python
studentlist = ["aryan","swastika","samaira","jiya","udita","bhavya"]
placedstudentlist = ["aryan","swastika","samaira"]

students = set(studentlist)
placed = set(placedstudentlist)

# Difference
not_placed = students - placed
print("Not placed students:", not_placed)

# Union
total_students = students | placed
print("Total students (union):", total_students)

# Intersection
common = students & placed
print("Placed students (intersection):", common)

# Symmetric difference
sym_diff = students ^ placed
print("Symmetric difference:", sym_diff)
```

```python
# Original set (1 to 10)
numbers = set(range(1, 11))

# Even and Odd sets
even = {x for x in numbers if x % 2 == 0}
odd = {x for x in numbers if x % 2 != 0}

print("Numbers:", numbers)
print("Even numbers:", even)
print("Odd numbers:", odd)

print("\n--- Set Operations ---")

# 1️⃣ isdisjoint
print("Even and Odd are disjoint?:", even.isdisjoint(odd))

# 2️⃣ subset
print("Even is subset of numbers?:", even.issubset(numbers))
print("Odd is subset of numbers?:", odd.issubset(numbers))

# 3️⃣ superset
print("Numbers is superset of even?:", numbers.issuperset(even))
print("Numbers is superset of odd?:", numbers.issuperset(odd))
```

```python
registration = 202400
d_students = {}

while True:
    choice = int(input("\n1. Register\n2. Search\n3. List all\n4. Exit\n5. Update\nEnter choice: "))

    # 1️⃣ Register
    if choice == 1:
        registration += 1
        reg_no = registration

        name = input("Name: ")
        gender = input("Gender: ")
        phone = input("Phone number: ")

        d_students[reg_no] = [name, gender, phone]

        print(f"Registration done. Your number is: {reg_no}")

    # 2️⃣ Search
    elif choice == 2:
        search_no = int(input("Enter registration number: "))

        if search_no in d_students:
            print("Details:", d_students[search_no])
        else:
            print("Student not found")

    # 3️⃣ List all
    elif choice == 3:
        if not d_students:
            print("No students registered yet")
        else:
            for reg_no, details in d_students.items():
                print(reg_no, "->", details)

    # 4️⃣ Exit
    elif choice == 4:
        print("Exiting program...")
        break

    # 5️⃣ Update
    elif choice == 5:
        reg_no = int(input("Enter registration number to update: "))

        if reg_no in d_students:
            print("Old details:", d_students[reg_no])

            name = input("New name: ")
            gender = input("New gender: ")
            phone = input("New phone: ")

            d_students[reg_no] = [name, gender, phone]

            print("Details updated successfully!")
        else:
            print("Student not found")

    # ❌ Invalid
    else:
        print("Invalid choice")
```
