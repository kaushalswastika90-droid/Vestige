```python
'''read a string and find the number of unique characters in it
'''

data="cranes varsity bangalore"
unique=set(data)

print("unique characters number: ",len(unique))
print("unique characters list: ",(unique))

data=list(data)
for ele in unique:
  data.remove(ele)

print("repeated characters: ", set(data))
```

```python
#3. Multiply All Numbers using *args:
def multiply(*args):
    result = 1
    for i in args:
        result *= i
    return result

nums = list(map(int, input().split()))
print(multiply(*nums))
```

```python
"""4. Concatenate Strings using *args:
Input: Hello World Python
Output: HelloWorldPython"""

def concatenate(*args):
    result = ""
    for i in args:
        result += i
    return result

strings = input().split()
print(concatenate(*strings))
```

```python
"""5. Print Info using **kwargs:
Input: name=Aniket age=22 city=Mumbai
Output:
name: Aniket
age: 22
city: Mumbai"""

def display(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

data = input().split()

info = {}
for item in data:
    k, v = item.split("=")
    info[k] = v

display(**info)
```

```python
"""19. Marksheet using **kwargs:
Input: math=90 sci=88 eng=92
Output:
math: 90
sci: 88
eng: 92"""
def display(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
data=input().split()
info={}
for item in data:
  k,v=item.split("=")
  info[k]=int(v)
display(**info)
```

```python
"""20. Reduce Maximum:
Input: 6 9 3 12 5
Output: 12"""
from functools import reduce

lst = list(map(int, input().split()))
result = reduce(lambda x, y: x if x > y else y, lst)

print(result)
```
