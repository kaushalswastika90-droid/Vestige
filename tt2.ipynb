```python
"""1. Reverse a List Problem Statement: Given a list of integers, reverse the list.
Input Format:- First line: integer n (size of list)-
Second line: n integers separated by space
Output Format:- Print the reversed list as space-separated integers."""
2

n = int(input("Enter size: "))
lst = list(map(int, input("Enter elements: ").split()))

reversed_list = lst[::-1]

print(*reversed_list)
```

```python
"""2. Maximum Element in a List Problem Statement: Find the maximum element in the list.
Input Format:- First line: integer n-
Second line: n integers
Output Format:- Single integer (maximum element)."""
n = int(input("Enter size: "))
lst = []

for i in range(n):
    num = int(input("Enter elements: "))
    lst.append(num)

max_val = lst[0]

for ch in lst:
    if ch > max_val:
        max_val = ch

print("max value",max_val)
```
